```python
import pandas as pd
import re

# Load CSV
df = pd.read_csv("DATASET/multilabel_BILSTM_BERT.csv")

import re

def clean_version(v):
    if pd.isna(v):
        return None
    
    v = str(v)

    # remove escape unicode
    v = v.replace("\\u003e", ">").replace("\\u003c", "<")

    # remove quotes
    v = v.replace('"', '').replace("'", "")

    # remove spaces
    v = re.sub(r"\s+", "", v)

    return v

# Hàm extract pragma version
def extract_solidity_version(source_code):
    if pd.isna(source_code):
        return None

    # Regex bắt pragma solidity
    match = re.search(r'pragma\s+solidity\s+([^;]+);', source_code)
    
    if match:
        return match.group(1).strip()
    return None

# Áp dụng
df['solidity_version'] = df['sourcecode'].apply(extract_solidity_version)
#df['solidity_version'] = df['solidity_version'].apply(clean_version)

# Xem kết quả
print(df['solidity_version'].value_counts().to_string())
```
